```python
import win32gui

toplist, winlist = [], []
def enum_cb(window_handle, results):
    winlist.append((window_handle, win32gui.GetWindowText(window_handle)))
win32gui.EnumWindows(enum_cb, toplist)

winlist
```

```python
from PIL import ImageGrab
import time

window = [(hwnd, title) for hwnd, title in winlist if 'valorant' in title.lower()]
# just grab the hwnd for first window matching title
valorant_window = window[0]
valorant_window_handle = valorant_window[0]

# win32gui.SetForegroundWindow(hwnd)
bbox = win32gui.GetWindowRect(valorant_window_handle)
```

```python
from ultralytics import YOLO

# model = YOLO('yolov8n.pt')
model = YOLO('best.pt')
```

```python
import pyautogui
from IPython.display import Image

pyautogui.PAUSE = 0
pyautogui.FAILSAFE = 0

# detect key press to exit loop
import keyboard

frame = 0
while True:
  if keyboard.is_pressed('q'):
    break

  print(pyautogui.position())
  # img = ImageGrab.grab(bbox)
  x0 = 200
  y0 = 300
  img = ImageGrab.grab((x0, y0, 900, 800))
  results = model.predict(source=img, conf=0.25)
  # print('Detected boxes: ', results[0].boxes.cls.shape)

  for result in results:
    result.save(filename=f'runs/detect/predict7/{frame}.jpg')
    for box in result.boxes.xyxy:
      x1, y1, x2, y2 = box
      x, y = x0 + (x1 + x2) / 2, y0 + (y1 + y2) / 2
      # print(x, y)
      pyautogui.moveTo(x, y)
      pyautogui.click(button='left')
  frame += 1
  time.sleep(0.1)
```
